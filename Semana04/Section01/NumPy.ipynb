**Transformando Array de 1D em 2D**

```python
a = np.array([1,2,3])
print(a.ndim)
print(a.shape)
```

```python
a2 = a[np.newaxis,:]
print(a2.shape)
print(a2.ndim)
print(a2)
```

```python
a22 = a[:,np.newaxis]
print(a22.shape)
print(a22.ndim)
print(a22)
```
