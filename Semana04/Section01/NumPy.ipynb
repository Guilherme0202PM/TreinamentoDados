**Transformando Array de 1D em 2D**

```python
a = np.array([1,2,3])
print(a.ndim)
print(a.shape)
```

```python
a2 = a[np.newaxis,:]
print(a2.shape)
print(a2.ndim)
print(a2)
```

```python
a22 = a[:,np.newaxis]
print(a22.shape)
print(a22.ndim)
print(a22)
```

**Concatenar Arrays**

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

c = np.concatenate((a,b))
d = np.concatenate((b,a))

print(c)
print(d)
```

**Consultar itens de um Arrays**

**Gerar amostras aleatórias**

**Diferença entre Arrays e Listas**

```python
a = np.array([1,2,3,4,5,6,7,8])
print("Esse é o array 'a':",a)
print("Esse é o tipo de 'a':", type(a))
print('-----------------------------------------')
lista_a = [1,2,3,4,5,6,7,8]
print("Essa pe a 'lista_a':", lista_a)
print("Esse é o tipo de 'lista_a':", type(lista_a))
```

```python
#arrays não permitem tipos de dados distintos:
a = np.array([1,'Daniel',2,3,4,5,6,7,8])
print(a)
print(type(a[0]))
```

```python
#já as listas sim:
lista_a = [1,'Daniel',2,3,4,5,6,7,8]
print(a)
print(type(lista_a[0]))
```
