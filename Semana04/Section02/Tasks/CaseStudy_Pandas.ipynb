**Calcule:**
* a. TP (real malign, previsto malign)
* b. TN (real benign, previsto benign)
* c. FP (real benign, previsto malign)
* d. FN (real malign, previsto benign)

```python
TP = ((data["real_class"] == "malign") &
      (data["predicted_class"] == "malign")).sum()
```

```python
TN = ((data["real_class"] == "benign") &
      (data["predicted_class"] == "benign")).sum()
```

```python
FP = ((data["real_class"] == "benign") &
      (data["predicted_class"] == "malign")).sum()
```

```python
FN = ((data["real_class"] == "malign") &
      (data["predicted_class"] == "benign")).sum()
```

```python
TP, TN, FP, FN
```
