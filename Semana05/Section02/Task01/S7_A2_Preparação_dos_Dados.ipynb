```python
import pandas as pd

# URL do dataset no GitHub
url = "https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_2017.csv"

# Carregar dados
dados = pd.read_csv(url)

```

```python
dados.head(5)
```

```python
dados.info()
```

```python
dados.describe(include='all')
```

```python
# Verificar quantidade de nulos por coluna
print(dados.isnull().sum())
```

```python
dados['winner_entry'].describe()
```

```python
dados['winner_entry'].unique()
```

```python
# Apagar linhas onde um atributo específico é nulo
dados.dropna(subset=['loser_rank'], inplace=True)
```

```python
dados.dropna(axis=1, how='all')
```

```python
dados['loser_entry'].unique()
```

```python
# Substituir nulos por um valor fixo (ex: 'x')
dados['loser_entry'].fillna('x', inplace=True)
```

```python
dados['loser_entry'].unique()
```

```python
# Substituir nulos pela média da coluna
media_altura = dados['winner_ht'].mean()
dados['winner_ht'].fillna(media_altura, inplace=True)
```

```python
# Projetar o valor do registro anterior para o nulo seguinte (Forward Fill)
dados['minutes'].fillna(method='ffill', inplace=True)
```
