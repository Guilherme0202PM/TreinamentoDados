2.Use o dataset escolhido e mostre por meio das representações gráficas (histogramas) as principais características. Procure usar as funções apresentadas nesta atividade.

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Class', data=df)
plt.title('Distribuição das Classes de Vinho')
plt.xlabel('Classe')
plt.ylabel('Quantidade')
plt.show()

# Gráfico da quantidade de amostras por classe de vinho
```

```python
classes = df['Class'].unique()

for coluna in df.columns[1:]:
    fig, eixos = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
    fig.suptitle(f'Histograma de {coluna} por Classe', fontsize=16)

    for i, classe in enumerate(classes):
        sns.histplot(df[df['Class'] == classe][coluna], bins=15, kde=True, ax=eixos[i])
        eixos[i].set_title(f'Classe {classe}')
        eixos[i].set_xlabel(coluna)
        eixos[i].set_ylabel('Frequência')

    plt.tight_layout(rect=[0, 0.05, 1, 0.9])
    plt.show()
```

```python
# Verifica se há valores nulos no DataFrame
df.loc[df.isnull().any(axis=1)]

#Não há valores nulos no dataset
# df.isnull().sum() faz a mesma coisa realizando a soma de valores nulos por coluna
```

```python
# Sustitui os valores NaN por valores zeros (0s):
df.fillna(0)

# Sustitui os valores NaN pela média da coluna:
df.fillna(df.mean())

# Sustitui os valores NaN pela mediana da coluna:
df.fillna(df.median())
```

5. Compare os resultados obtidos usando um gráfico box plot. Que atributos possuem distribuições mais simétricas? Há a presença de valores extremos, denominados de outliers, para os atributos?

```python
for coluna in df.columns[1:]:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Class', y=coluna, data=df)
    plt.title(f'Boxplot de {coluna} por Classe')
    plt.xlabel('Classe')
    plt.ylabel(coluna)
    plt.show()
```

6. No dataset escolhido selecione intervalos de amostragem (10%, 30% e 50% - Progressiva) e faça análises sobre os dados em relação as métricas estatísticas. O que você pode observar sobre os subconjuntos em relação as métricas estatísticas com a nova composição dos dados?

```python
amostragens = [0.1, 0.3, 0.5, 0.75, 1.0]

for amostra in amostragens:
    print(f"\n{'-'*25} Tamanho da amostra: {amostra*100}% {'-'*25}\n")
    n = int(len(df) * amostra)
    df_amostra = df.sample(n=n, random_state=42)
    print(df_amostra.describe())
```
